- <font color='orange'>Rescaling:</font>
    - Rescala para o intervalo entre 0 e 1;
    - Distribuição não Gaussianas;

- <font color='orange'>Transformação:</font>
    - Conversão de Features Categóricas para Númericas (Encoding);
    - Transformação de Natureza;
