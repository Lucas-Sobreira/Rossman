### Características do Erro MAE

<ul>a. Atribui peso igual para todos os erros;</ul>

## <font color='yellow'>MAPE (Mean Absolute Percentage Error)</font>

```python
PATH = 'D:\\Backup\\Comunidade DS\\Cursos\\DS_Producao\\images\\MAPE.png'
Image(filename= PATH, width=750, height=380)
```

### Características do MAPE

<ul>a. Mostra o quão longe a predição está do valor real, na média, em porcentagem;</ul>
<ul>b. Muito usado na divulgação de Resultados;</ul>
<ul>c. Não pode ser usado se a Variável Resposta contém 0;</ul>

### Exemplo - Diferença entre Modelos

```python
# Erro Atual
MAE_atual = 1100
MAPE_atual = 17/100

# Erro Modelo
MAE_modelo = 665
MAPE_modelo = 10/100

# Melhoria
MAE_melhoria = MAE_atual - MAE_modelo
MAPE_melhoria = MAPE_atual - MAPE_modelo

print(f'MAE melhoria: {MAE_melhoria}')
print(f'MAPE melhoria: {round(MAPE_melhoria*100, 2)}%')
```
